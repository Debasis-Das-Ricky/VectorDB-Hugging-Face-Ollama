1. What is a Vector Database (VectorDB) and how is it different from traditional databases?

 Ans. A Vector Database (VectorDB) is a specialized database designed to store, index, and search high dimensional numerical vectors, typically embeddings produced by machine learning models from text, images, audio, or other unstructured data. It enables similarity search based on distance metrics (like cosine similarity or Euclidean distance) so that queries retrieve results with similar meaning rather than exact keyword matches. VectorDBs are therefore widely used in generative AI workflows such as RAG, semantic search, recommendation systems, and intelligent chatbots.

 In contrast, traditional databases (ex. relational databases) store mainly structured data in tables with rows and columns and are optimized for exact lookups, joins, and transactional workloads using SQL and B tree or hash indexes. Their queries typically use equality and range conditions and return precisely matching records, making them ideal for use cases like banking transactions, inventory management, and reporting. Traditional databases can store embeddings as numbers but are not optimized for high dimensional similarity search at scale, often resulting in higher latency and poorer relevance for AI applications compared to a dedicated VectorDB.



2. Explain the various types of VectorDBs available and describe their suitability for different use cases.

 Ans. Different types of VectorDBs focus on where they store data, how they index vectors, and how they are deployed. Each type suits different scale, latency, and operational requirements in data science and generative AI applications.

 By storage model:

  i) In memory VectorDBs: Store vectors mainly in RAM, give extremely low latency queries, but are limited by memory size and cost. Best for real time recommendation, interactive chatbots, and high frequency RAG queries where response time is critical.

  ii) Disk based VectorDBs: Keep vectors on disk, can handle very large datasets but with higher latency than pure in memory systems. Suitable for large historical corpora, log search, analytics, and RAG over big document collections where ultra fast responses are less critical.

  iii) Hybrid VectorDBs: Combine RAM for "hot" data with disk for "cold" data to balance speed and capacity. Good for production RAG and search systems that must serve both fresh, frequently accessed data and huge archives efficiently.

 By deployment / product style:

  i) Standalone managed VectorDBs (ex. Pinecone): Fully managed cloud services focused only on vector search, offering high scalability, low latency indexes, and simple APIs. Well suited for SaaS AI products and teams that want to avoid infrastructure management and scale to billions of vectors.

  ii) Open source VectorDBs (ex. Milvus, Weaviate, Qdrant): Self hosted or cloud hosted, provide flexible integration, custom configs, and no license lock in. Appropriate for organizations needing on prem deployment, cost control, or deep customization of indexing and hardware (CPU/GPU).

  iii) Traditional DBs with vector extensions (ex PostgreSQL with pgvector, MongoDB Atlas Vector Search, Redis Vector Search): Add vector indexes on top of existing relational/NoSQL systems. Best when you want hybrid workloads (transactions + vector search) in one system, moderate scale, and easier integration with existing data models.

  By indexing technique:

  i) Tree based indexes (KD tree, R tree etc.): Work well for lower dimensional data and smaller datasets, but scale poorly to very high dimensions. Useful for classical ML or spatial search where dimensions are limited (ex geospatial vectors, simple feature vectors).

  ii) Hashing based indexes (LSH etc.): Use hashing of vectors for very fast approximate lookup on huge datasets. Suitable for web scale similarity search where slight loss in accuracy is acceptable for big gains in speed.

  iii) Quantization based indexes (ex. IVF, PQ): Compress vectors into compact codes to save memory and accelerate approximate nearest neighbor search. Ideal for billion scale embedding collections (large RAG corpora, recommendation catalogs) running on limited hardware.



3. Why is Chroma DB considered important in the context of AI/ML projects? Describe its key features.

 Ans. Chroma DB is important because it gives AI/ML projects an easy, open source "memory layer" to store embeddings and do fast semantic search, which is essential for RAG, chatbots, and recommendation systems. It reduces boilerplate (storage + embeddings + search) and lets developers move quickly from prototype to production with simple APIs.

 Why it is important:

  i) AI native vector database: Chroma is designed specifically for storing and querying vector embeddings (from text, images, etc.), so it fits modern LLM and GenAI workflows very well.

  ii) Powers RAG and “AI memory”: It acts as the knowledge base for retrieval augmented generation, grounding LLMs with relevant context and helping reduce hallucinations in real systems.

  iii) Fast and practical for developers: It offers quick similarity search over embeddings with minimal setup, so you can give your AI app “memory” in a few lines of code.

 Key features:

  i) Open source and developer friendly: Fully open source, with simple Python and JavaScript clients and integrations with tools like LangChain, LlamaIndex, and Streamlit.

  ii) Efficient vector storage and search: Stores embeddings as vectors, uses in memory plus optional persistent storage, and supports fast nearest neighbor search for large text/image corpora.

  iii) Flexible storage modes: Supports temporary (ephemeral) storage for prototyping and persistent storage for production, so you can start small and later keep long term data safely.

  iv) Model and tool integration: Works smoothly with popular embedding providers (OpenAI, Google, Cohere, Hugging Face) or custom models, making it easy to plug into existing ML pipelines.

  v) Rich retrieval capabilities: Provides vector similarity search, metadata filtering, and multi modal retrieval, which are crucial for realistic search, RAG, and recommender systems.

  vi) Scalability and performance: Uses efficient indexing (ex. HNSW) and can scale horizontally or via client server mode, keeping search fast even as datasets grow.

```python
# 6)  Install and set up Chroma DB, and insert sample vector data for semantic search.

# Step 1: Install ChromaDB (run this in terminal first)
!pip install chromadb

# Step 2: Import and set up ChromaDB client (persistent mode)
import chromadb
import numpy as np

# Create persistent client (data saved to ./chroma_db folder)
client = chromadb.PersistentClient(path="./chroma_db")

# Step 3: Create or get a collection for storing vectors
collection = client.get_or_create_collection(name="sample_vectors")

# Step 4: Prepare sample data for semantic search
# Sample texts (for context)
documents = [
    "The quick brown fox jumps over the lazy dog.",
    "A fast brown fox leaps over a sleeping dog.",
    "Python is a popular programming language.",
    "Machine learning uses algorithms to learn from data."
]

# Sample vectors (simple embeddings, in real use: from sentence-transformers or OpenAI)
vectors = [
    np.array([0.1, 0.8, 0.2, 0.9]),  # close to "fox jumps dog"
    np.array([0.12, 0.78, 0.22, 0.88]),  # very similar to first
    np.array([0.7, 0.3, 0.6, 0.1]),  # "Python programming"
    np.array([0.65, 0.35, 0.62, 0.12])  # "ML data"
]

# Unique IDs and metadata
ids = ["doc1", "doc2", "doc3", "doc4"]
metadatas = [
    {"topic": "animals"},
    {"topic": "animals"},
    {"topic": "programming"},
    {"topic": "ml"}
]

# Step 5: Insert data into collection
collection.add(
    documents=documents,
    embeddings=vectors,
    ids=ids,
    metadatas=metadatas
)

print("Sample vectors inserted! Count:", collection.count())

# Step 6: Test semantic search
query_vector = np.array([0.11, 0.79, 0.21, 0.89])  # close to fox docs
results = collection.query(
    query_embeddings=[query_vector],
    n_results=2
)

print("\nSemantic search results:")
print("IDs:", results['ids'][0])
print("Distances:", results['distances'][0])
print("Documents:", results['documents'][0])
```

```python
# 7)  Demonstrate how to download and fine-tune a Hugging Face model for a text generation task.

# Step 1: Install required libraries
#!pip install -q transformers datasets accelerate peft bitsandbytes
#!pip install -q evaluate rouge_score

# Step 2: Import libraries
import torch
from datasets import load_dataset, Dataset
from transformers import (
    AutoTokenizer, AutoModelForCausalLM, TrainingArguments, Trainer,
    DataCollatorForLanguageModeling, BitsAndBytesConfig
)
from peft import LoraConfig, get_peft_model, TaskType
import evaluate
import numpy as np
import requests

# Step 3: Download pre-trained model and tokenizer
model_name = "microsoft/DialoGPT-medium"  # Good for text generation (dialogue)
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto",  # Auto GPU/CPU
    trust_remote_code=True
)

# Add pad token if missing
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

print(f"✅ Model '{model_name}' downloaded! Params: {model.num_parameters():,}")

# Step 4: Load and prepare dataset (tiny Shakespeare for demo)
# Download the raw text content for tiny_shakespeare
url = "https://raw.githubusercontent.com/karpathy/makemore/master/input.txt"
response = requests.get(url)

# Read content and create dataset from lines
lines = response.text.splitlines()
# Filter out empty lines
lines = [line for line in lines if line.strip()]

# Create a Dataset object from the list of lines
dataset = Dataset.from_dict({"text": lines})

# Safely select a subset for fast training
num_samples_to_select = 1000
if len(dataset) > num_samples_to_select:
    dataset = dataset.shuffle(seed=42).select(range(num_samples_to_select))
else:
    print(f"Warning: Dataset has only {len(dataset)} examples, using all for training.")
    dataset = dataset.shuffle(seed=42)

def tokenize_function(examples):
    # Tokenize for causal LM (text generation)
    tokenized = tokenizer(
        examples["text"],
        truncation=True,
        padding=False,
        max_length=128,
        return_tensors="pt"
    )
    # Labels = input_ids for autoregressive training
    tokenized["labels"] = tokenized["input_ids"].clone()
    return tokenized

tokenized_dataset = dataset.map(tokenize_function, batched=True, remove_columns=dataset.column_names)
train_dataset = tokenized_dataset.train_test_split(test_size=0.1)["train"]
eval_dataset = tokenized_dataset.train_test_split(test_size=0.1)["test"]

print(f"✅ Dataset ready: {len(train_dataset)} train, {len(eval_dataset)} eval samples")

# Step 5: Apply LoRA for efficient fine-tuning (only trains ~1% of params)
peft_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=16,  # Rank
    lora_alpha=32,
    lora_dropout=0.1,
    target_modules=["c_attn", "c_proj"]  # GPT-2 specific
)
model = get_peft_model(model, peft_config)
model.print_trainable_parameters()  # Shows % trainable params

# Step 6: Setup training
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

# Metrics (ROUGE for generation quality)
rouge = evaluate.load("rouge")

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)
    result = rouge.compute(predictions=decoded_preds, references=decoded_labels)
    return {"rouge2": result["rouge2"]}

# Training args (fast for demo)
training_args = TrainingArguments(
    output_dir="./fine_tuned_dialogpt",
    overwrite_output_dir=True,
    num_train_epochs=1,  # Increase for better results
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=4,
    warmup_steps=100,
    logging_steps=50,
    eval_strategy="steps",
    eval_steps=200,
    save_steps=500,
    save_total_limit=2,
    report_to=None,  # No wandb
    fp16=True,
    dataloader_pin_memory=False
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# Step 7: Fine-tune!
print("🚀 Starting fine-tuning...")
trainer.train()
trainer.save_model("./fine_tuned_dialogpt")
tokenizer.save_pretrained("./fine_tuned_dialogpt")

print("✅ Fine-tuning complete! Model saved.")

# Step 8: Test the fine-tuned model
test_prompt = "To be, or not to be"
inputs = tokenizer(test_prompt, return_tensors="pt").to(model.device)
with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=50,
        do_sample=True,
        temperature=0.7,
        pad_token_id=tokenizer.eos_token_id
    )
response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(f"\nGenerated text:\n'{response}'")
```

```python
# 8) Create a custom LLM using Ollama and Llama2, and run it locally for basic text prompts.

# === COMPLETE LOCAL SETUP: Custom Llama2 LLM with Ollama + Python ===
# Runs 100% locally. Perfect for your data science projects.

# PREREQUISITES (run these FIRST in your terminal):
# 1. Install Ollama: https://ollama.com/download (Linux/Mac/Windows)
#    - Linux: curl -fsSL https://ollama.com/install.sh | sh
#    - Then start: ollama serve  (keep running in background)
#
# 2. Pull base Llama2 model: ollama pull llama2  (3.8B params, ~2GB)
#
# 3. Install Python client: pip install ollama

# Step 1: Import Ollama Python client
# !pip install ollama
import ollama

print("✅ Ollama client ready! Starting custom LLM...")

# Step 2: Create CUSTOM model from Llama2 with system prompt (Modelfile)
# This makes it a "custom LLM" tailored for data science tasks
custom_modelfile = '''
FROM llama2

# Custom system prompt: Make it a Data Science Assistant (your background!)
SYSTEM """
You are DataScienceBot, a helpful assistant for MSc Data Science students.
You specialize in Python, ML/DL, object detection (Detectron2), vector DBs,
and GenAI. Give practical code examples, explain concepts simply, and suggest
next steps for projects like fraud detection or RAG systems.
Keep answers concise, use bullet points, and focus on Bhubaneswar/India context
when relevant.
"""

# Optional: Set parameters for better generation
PARAMETER temperature 0.7
PARAMETER top_p 0.9
'''

# Create the custom model (runs once, ~10-30 sec)
# The `modelfile` content needs to be passed directly as a string, not as a keyword argument.
ollama.create('datasciencebot', custom_modelfile)
print("✅ Custom 'datasciencebot' model created from Llama2!")

# Step 3: Test with basic text prompts (interactive chat)
print("\n🤖 Chat with your custom DataScienceBot (type 'quit' to exit):")

messages = [
    {'role': 'system', 'content': 'You are DataScienceBot.'}
]

while True:
    user_prompt = input("\nYou: ")
    if user_prompt.lower() in ['quit', 'exit', 'q']:
        break

    messages.append({'role': 'user', 'content': user_prompt})

    # Generate response using custom model
    response = ollama.chat(
        model='datasciencebot',
        messages=messages,
        stream=False  # Set True for streaming
    )

    bot_reply = response['message']['content']
    print(f"\nDataScienceBot: {bot_reply}")

    messages.append({'role': 'assistant', 'content': bot_reply})

print("\n👋 Session ended. List models: ollama list | Delete: ollama rm datasciencebot")
```

```python
# 9)  Implement a basic RAG (Retrieval-Augmented Generation) system using Ollama with Llama3.

# === COMPLETE LOCAL RAG SYSTEM: Ollama + Llama3 + ChromaDB ===
# 100% LOCAL, no APIs. Perfect for your GenAI data science projects!

# PREREQUISITES (Terminal commands - run FIRST):
# 1. Install Ollama: https://ollama.com/download
# 2. Pull models:
#    ollama pull llama3.2          # LLM (3B, fast)
#    ollama pull nomic-embed-text   # Embeddings
# 3. pip install chromadb sentence-transformers numpy ollama

# Step 1: Imports
import chromadb
from chromadb.utils import embedding_functions
import ollama
import numpy as np
from sentence_transformers import SentenceTransformer  # Fallback embedding

print("✅ RAG imports ready!")

# Step 2: Setup ChromaDB + Embedding function (Ollama native)
client = chromadb.PersistentClient(path="./rag_db")
ollama_ef = embedding_functions.OllamaEmbeddingFunction(
    model_name="nomic-embed-text"
)

# Create collection
collection = client.get_or_create_collection(
    name="data_science_docs",
    embedding_function=ollama_ef
)

# Step 3: Sample documents (your data science notes/projects)
documents = [
    "Vector databases store high-dimensional embeddings for semantic search. Used in RAG for LLMs.",
    "Detectron2 is Facebook's object detection library. Supports Mask R-CNN, RetinaNet for CV projects.",
    "ChromaDB is an open-source vector DB. Easy Python API for local AI apps and prototyping.",
    "RAG combines retrieval from vector store + generation by LLM to reduce hallucinations.",
    "LoRA fine-tuning adapts large models efficiently by training small adapter matrices.",
    "Ollama runs LLMs locally. Pull models like Llama3 and use Python client for inference."
]

# Step 4: Add documents to vector store
ids = [f"doc_{i}" for i in range(len(documents))]
collection.add(
    documents=documents,
    ids=ids
)
print(f"✅ Inserted {collection.count()} docs into ChromaDB!")

# Step 5: RAG Pipeline Function
def rag_query(question, n_results=2):
    # 5a: Retrieve relevant docs
    results = collection.query(
        query_texts=[question],
        n_results=n_results
    )

    context = "\n".join(results['documents'][0])

    # 5b: Generate with Llama3 + retrieved context
    prompt = f"""You are a helpful Data Science assistant. Answer based ONLY on this context:

CONTEXT:
{context}

QUESTION: {question}

ANSWER:"""

    response = ollama.generate(
        model="llama3.2",
        prompt=prompt,
        options={
            'temperature': 0.1,  # Low for factual answers
            'num_predict': 200
        }
    )

    return {
        'answer': response['response'],
        'sources': results['documents'][0],
        'ids': results['ids'][0]
    }

# Step 6: Interactive RAG Chat
print("\n🤖 RAG Data Science Assistant (type 'quit' to exit):")
while True:
    question = input("\nYou: ")
    if question.lower() in ['quit', 'exit', 'q']:
        break

    rag_result = rag_query(question)

    print("\n📄 Answer:")
    print(rag_result['answer'])
    print("\n📚 Sources:")
    for i, doc in enumerate(rag_result['sources']):
        print(f"- {rag_result['ids'][i]}: {doc[:100]}...")

    print("-" * 80)

print("\n👋 RAG session ended. Data persists in ./rag_db")
```
